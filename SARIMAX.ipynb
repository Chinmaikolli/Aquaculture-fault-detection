```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.interpolation import shift
%matplotlib inline
```

```python
import pandas as pd
```

```python
df = pd.read_csv('fault.csv')
print (df.head) 
```

```python
from datetime import datetime, timedelta

# get current time
now = datetime.now()
print ("Today's date: ", str(now))
#add 15 days to current date

lst = []
for i in range(len(df['PH'])):
    future_date_after_10sec= now + timedelta(seconds = 60)
    now = future_date_after_10sec
    lst.append(now)
    print(now)
   
```

```python
df['time'] = pd.DataFrame(lst)
```

```python
df
```

```python
df_temp = df[['Temperature','time']] 
df_temp = df_temp.set_index('time')
df_temp  
```

```python

import statsmodels.api as sm

```

```python
mod = sm.tsa.statespace.SARIMAX(df_temp,
                                order=(1, 0, 1),
                                seasonal_order=(1, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
pred = results.get_prediction(start=pd.to_datetime('2020-05-13 19:52:00.121360'), dynamic=False)
pred_ci = pred.conf_int()
ax = df_temp['2020-05-13 19:52:00.121360':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='r', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Temp')
plt.ylim(1, 40)
plt.legend()
plt.show()
```
