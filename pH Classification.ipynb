```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("NEW.csv")
```

```python
df.head()
```

```python
import random
df.dropna(inplace = True)
df
```

```python
df_work = df
df_work['Label']=0
df_work
```

```python
df_work= df_work[['PH','Label']]
```

```python
df_work.head()
```

```python
df_work.describe()
```

```python
faults = []
for i in range(400):
    faults.append(random.uniform(0, 1))
    
faults = [i*14 for i in faults]
print(faults)

    
```

```python
for index,i in enumerate(df_work['PH']):
    if(index%100 ==0):
        df_work.at[index,'PH'] = faults[index%400]
        df_work.at[index,'Label']= 1
        
        
diff = pd.DataFrame(df_work.diff())


diff_1 =  pd.DataFrame(diff.loc[:,'PH']<-1)
print(diff_1) 


for index,i in enumerate(diff_1):        
    if(i==True):
        df_work.at[index, 'PH'] = 1
    else:
        df_work.at[index, 'PH'] = 1
```

```python
df_work
```

```python
x = df_work[['PH']]
y = df_work[['Label']]
            
from sklearn.model_selection import train_test_split
            
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)

y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
import sklearn

clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)

print(accuracy_score(y_pred,y_test))
print(list(y_test[0:400]))
print(y_pred[0:400])
#print(X_test[0:400])

print(sklearn.metrics.classification_report(y_test,y_pred))
print(sklearn.metrics.roc_auc_score(y_test,y_pred))
```

```python
import sklearn as sk
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X = np.array(df_work['PH'])
y = np.array(df_work['Label'])
X = X.reshape((-1,1))
LR = LogisticRegression(random_state=1, solver='lbfgs').fit(X,y)



print(LR.intercept_)
print(LR.coef_)

lr = LR.predict(x_test)
print(accuracy_score(y_test,lr))

print(sklearn.metrics.classification_report(y_test,lr ,labels=np.unique(lr)))
print(sklearn.metrics.roc_auc_score(y_test,lr))
metrics.f1_score(y_test, lr, average='weighted', labels=np.unique(lr))
```

```python
from sklearn import svm
from sklearn import linear_model 



SVM = svm.LinearSVC(max_iter=5000)
SVM.fit(x_train, y_train)

s = SVM.predict(x_test)


print(accuracy_score(s,y_test))
print(sklearn.metrics.roc_auc_score(y_test,lr))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=1, max_depth=8, random_state=1)
RF.fit(x_train, y_train)
rf = RF.predict(x_test)
print(accuracy_score(rf,y_test))
print(sklearn.metrics.classification_report(y_test,rf))
print(sklearn.metrics.roc_auc_score(y_test,rf))
```
